```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
athlete = r'C:\Users\Riaz\Desktop\pandas\Olympics\athlete_events.csv'
athlete = pd.read_csv(athlete)
athlete
```

```python
region = r'C:\Users\Riaz\Desktop\pandas\Olympics\noc_regions.csv'
region = pd.read_csv(region)
region
```

# Join two data frame...

```python
athlete_region = athlete.merge(region, how = 'right', on = 'NOC')
athlete_region.head(10)
```

### Rename the column name

```python
athlete_region.rename(columns={'region': 'Region', 'notes':'Notes'}, inplace = True)
```

```python
athlete_region
```

### statisticale values...

```python
athlete_region.describe()
```

```python
athlete_region.drop('Games',axis= 1, inplace=True)
```

```python
athlete_region
```

### Details about China...

```python
athlete_region.query('Region== "China"').head(20)
```

### Top countries participating...

```python
top_10 = athlete_region['Region'].value_counts().sort_values(ascending= False).head(10)
top_10
```

### Ploting the graph...

```python


plt.figure(figsize=(12,6))
plt.title('Top 10 Contries')
plt.xlabel('Region')
plt.ylabel('Times')

sns.barplot(x = top_10.index, y = top_10, palette = 'Set2');
plt.show()
```

### Age distribution...

```python
plt.figure(figsize=(12,6))
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of participant')
plt.hist(athlete_region.Age, bins = np.arange(10,80,2), color = 'black', edgecolor = 'yellow')
```

### Summer sports olympic...

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_cout, labels= gender_cout.index, autopct='%1.1f%%', startangle= 150)
```

### total number of female participate in olympics...

```python
female_participate = athlete_region[(athlete_region.Sex == 'F') & (athlete_region.Season == 'Summer')][['Sex','Year']]
female_participate = female_participate.groupby('Year').count().reset_index()
female_participate
```

### Graph of female participant...

```python
female_participate = athlete_region[(athlete_region.Sex == 'F') & (athlete_region.Season == 'Summer')]
plt.figure(figsize=(20,6))
plt.title('Female Participation')
sns.countplot(x = 'Year',data = female_participate, palette='Paired')
```

### line graph of female participate over time...

```python
female_participate = female_participate.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,16))
female_participate.loc[:,'F'].plot()
plt.title("female participant over time")
```

### Gold Medal athlete...

```python
gold_medle = athlete_region[(athlete_region.Medal =='Gold')]
gold_medle
```

### Gold Medal for each country...

```python
gold_country = gold_medle.Region.value_counts().reset_index(name = 'Medal').head(10)
gold_country

sns.catplot(x = 'index', y = 'Medal', data = gold_country, kind = 'bar', palette="viridis", height= 8)

plt.title('Most Gold Medal Country')
plt.xlabel('Country')
plt.ylabel('Medal')
```

### Most Gold medal county in 2004... 

```python
gold_2004 = athlete_region[(athlete_region.Year == 2004) & (athlete_region.Medal == "Gold")].Region.value_counts().head()
gold_2004
```

```python
plt.figure(figsize=(12,6))
plt.title('Top 5 Contries achived gold in 2004')
plt.xlabel('Region')
plt.ylabel('Count')

sns.barplot(x = gold_2004.index, y = gold_2004, palette = "coolwarm");
plt.show()
```
